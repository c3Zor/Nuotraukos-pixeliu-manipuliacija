# Kuriam filtrÄ… kuriuo galima isskirti spalvas.
### sukuriam kiekvienai spalvai filtrÄ…. 
Toliau naudojam funkcija logical_and(). ir pritaikome filtÄ… nuotraukai.

```python
spalvos = imageio.imread('./wifire/tokoro.jpg')
apdirbimas = np.asarray(spalvos)

raudona   = apdirbimas[:, : ,0] < 170
zalia = apdirbimas[:, : ,1] < 220
melyna  = apdirbimas[:, : ,2] > 10

filtras = np.logical_and(raudona, zalia, melyna)
apdirbimas[filtras] = 0
plt.figure(figsize=(15,15))
plt.imshow(apdirbimas)
```
